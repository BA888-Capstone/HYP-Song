```python
import pandas as pd
import numpy as np
```

```python
df_start = pd.read_csv('/content/sound_track.csv')
df_start
```

```python
df = df_start.drop(columns = ['popularity','explicit','key','mode','duration_ms','Unnamed: 0','track_id','artists','album_name','track_name','time_signature','track_genre'])
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scale = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)
df_scale
```

```python
X = df_scale[['danceability','energy','loudness','tempo']]
speechiness = df_scale['speechiness']
acous = df_scale['acousticness']
ins = df_scale['instrumentalness']
liveness = df_scale['liveness']
val = df_scale['valence']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

# liveness

```python
X_train, X_test, y_train, y_test = train_test_split(X, liveness, test_size=0.2, random_state=0)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
# Evaluate the model
rmse = mean_squared_error(y_test, y_pred, squared = False)
print(f'Root Mean Squared Error on test data: {rmse:.2f}')

# What would the error be if measured on the training data?
train_rmse = mean_squared_error(y_train, model.predict(X_train), squared = False)
print(f'Root Mean Squared Error on training data: {train_rmse:.2f}')
```
